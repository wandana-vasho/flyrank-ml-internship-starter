Question: Does a page's average ranking position meaningfully predict its click-through rate, and can this relationship be used to prioritize which pages a content team reviews first?

## 3. Quick look at the data (2-3 real numbers)

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/wandana-vasho/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv")
df[['avg_position', 'ctr']].describe()
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
correlation = df['avg_position'].corr(df['ctr'])
avg_ctr_top10 = df[df['avg_position'] <= 10]['ctr'].mean()
avg_ctr_below10 = df[df['avg_position'] > 10]['ctr'].mean()

print(f"Correlation between avg_position and ctr: {correlation:.3f}")
print(f"Average CTR for pages in position 1-10: {avg_ctr_top10:.3f}")
print(f"Average CTR for pages beyond position 10: {avg_ctr_below10:.3f}")
```
