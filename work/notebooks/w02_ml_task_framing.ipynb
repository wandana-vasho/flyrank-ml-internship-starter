My lane (from ML-02): identifying pages that underperform on CTR relative to their search ranking position, to prioritize which ones get reviewed first.

Task type: Scoring / Ranking.

This isn't classification (I'm not sorting pages into fixed categories) or clustering (I'm not grouping similar pages together). It's a scoring problem: for each page, produce a continuous score representing "how much is this page underperforming its expected CTR for its position?" Pages are then ranked by that score, and the top of the ranked list becomes the review queue.

True target (ideal, but not directly available): whether a content review actually improved this page's CTR — this requires running an experiment and waiting for results, which isn't available in the starter data.

Proxy I'll use instead: the gap between a page's actual CTR and its expected CTR given its position. A large negative gap means a page is underperforming for where it ranks.

This is a proxy, not ground truth: a low score suggests a page worth reviewing, but doesn't guarantee a review will fix it.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/wandana-vasho/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv")

unit_of_analysis = df[['avg_position', 'ctr']].copy()
unit_of_analysis['expected_ctr'] = unit_of_analysis['avg_position'].apply(
    lambda p: 0.787 if p <= 10 else 0.263
)
unit_of_analysis['ctr_gap'] = unit_of_analysis['ctr'] - unit_of_analysis['expected_ctr']

unit_of_analysis.head(10)
```
